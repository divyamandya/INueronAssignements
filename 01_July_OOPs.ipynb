```python
5. What is the difference between class variables and instance variables in Python?
Answer: Class Variables: Shared among all instances. Changing them via the class name changes them for all instances, but changing them via an instance creates an instance variable of the same name for that particular instance.
Instance Variables: Unique to each instance. Changing them affects only the particular instance they belong to.
```

```python
class Dog:
    # Class variable
    species = "Canis familiaris"

    def __init__(self, name, age):
        # Instance variables
        self.name = name
        self.age = age

# Creating instances of the Dog class
dog1 = Dog("Buddy", 3)
dog2 = Dog("Milo", 5)

# Accessing class variable
print(dog1.species)  # Output: Canis familiaris
print(dog2.species)  # Output: Canis familiaris

# Accessing instance variables
print(dog1.name)  # Output: Buddy
print(dog1.age)   # Output: 3
print(dog2.name)  # Output: Milo
print(dog2.age)   # Output: 5

# Modifying class variable through the class name
Dog.species = "Canis lupus familiaris"
print(dog1.species)  # Output: Canis lupus familiaris
print(dog2.species)  # Output: Canis lupus familiaris

# Modifying class variable through an instance (creates an instance variable)
dog1.species = "Unknown"
print(dog1.species)  # Output: Unknown (instance variable)
print(dog2.species)  # Output: Canis lupus familiaris (class variable)

# The class variable itself remains unchanged
print(Dog.species)  # Output: Canis lupus familiaris
```

```python
7. For a library management system, you have to design the "Book" class with OOP
principles in mind. The “Book” class will have following attributes:
a. title: Represents the title of the book.
b. author: Represents the author(s) of the book.
c. isbn: Represents the ISBN (International Standard Book Number) of the book.
d. publication_year: Represents the year of publication of the book.
e. available_copies: Represents the number of copies available for checkout.

Answer:To design the Book class for a library management system using Object-Oriented Programming (OOP) principles, we need to encapsulate the attributes and provide methods to interact with the book objects. Here's how the Book class can be implemented in Python:
```

```python
class Book:
    def __init__(self, title, author, isbn, publication_year, available_copies):
        """
        Initialize a new Book instance.

        :param title: str, the title of the book
        :param author: str, the author(s) of the book
        :param isbn: str, the ISBN of the book
        :param publication_year: int, the year the book was published
        :param available_copies: int, the number of available copies for checkout
        """
        self.title = title
        self.author = author
        self.isbn = isbn
        self.publication_year = publication_year
        self.available_copies = available_copies

    def check_out(self):
        """
        Decrement the number of available copies by one if any copies are available.
        
        :return: bool, True if checkout is successful, False otherwise
        """
        if self.available_copies > 0:
            self.available_copies -= 1
            return True
        else:
            return False

    def return_book(self):
        """
        Increment the number of available copies by one.
        """
        self.available_copies += 1

    def display_book_info(self):
        """
        Display the information about the book, including its attributes and the number of available copies.
        
        :return: str, formatted book information
        """
        return (f"Title: {self.title}\n"
                f"Author: {self.author}\n"
                f"ISBN: {self.isbn}\n"
                f"Publication Year: {self.publication_year}\n"
                f"Available Copies: {self.available_copies}\n")
    
    def __str__(self):
        """
        Provide a string representation of the book.

        :return: str, formatted book information
        """
        return self.display_book_info()

# Example usage:
if __name__ == "__main__":
    book1 = Book("1984", "George Orwell", "1234567890", 1949, 5)
    print(book1)

    # Checking out a book
    if book1.check_out():
        print("Checked out successfully")
    else:
        print("No copies available")
    print(book1)

    # Returning a book
    book1.return_book()
    print("Book returned")
    print(book1)
```

```python
8. For a ticket booking system, you have to design the "Ticket" class with OOP
principles in mind. The “Ticket” class should have the following attributes:
a. ticket_id: Represents the unique identifier for the ticket.
b. event_name: Represents the name of the event.
c. event_date: Represents the date of the event.
d. venue: Represents the venue of the event.
e. seat_number: Represents the seat number associated with the ticket.
f. price: Represents the price of the ticket.
g. is_reserved: Represents the reservation status of the ticket.
The class also includes the following methods:
a. reserve_ticket(self): Marks the ticket as reserved if it is not already reserved.
b. cancel_reservation(self): Cancels the reservation of the ticket if it is already
reserved.
c. display_ticket_info(self): Displays the information about the ticket, including its
attributes and reservation status.
```

```python
class Ticket:
    def __init__(self, ticket_id, event_name, event_date, venue, seat_number, price):
        """
        Initialize a new Ticket instance.

        :param ticket_id: str, the unique identifier for the ticket
        :param event_name: str, the name of the event
        :param event_date: str, the date of the event
        :param venue: str, the venue of the event
        :param seat_number: str, the seat number associated with the ticket
        :param price: float, the price of the ticket
        """
        self.ticket_id = ticket_id
        self.event_name = event_name
        self.event_date = event_date
        self.venue = venue
        self.seat_number = seat_number
        self.price = price
        self.is_reserved = False

    def reserve_ticket(self):
        """
        Marks the ticket as reserved if it is not already reserved.

        :return: bool, True if the ticket was successfully reserved, False if it was already reserved
        """
        if not self.is_reserved:
            self.is_reserved = True
            return True
        return False

    def cancel_reservation(self):
        """
        Cancels the reservation of the ticket if it is already reserved.

        :return: bool, True if the reservation was successfully canceled, False if it was not reserved
        """
        if self.is_reserved:
            self.is_reserved = False
            return True
        return False

    def display_ticket_info(self):
        """
        Displays the information about the ticket, including its attributes and reservation status.

        :return: str, formatted ticket information
        """
        reservation_status = "Reserved" if self.is_reserved else "Available"
        return (f"Ticket ID: {self.ticket_id}\n"
                f"Event Name: {self.event_name}\n"
                f"Event Date: {self.event_date}\n"
                f"Venue: {self.venue}\n"
                f"Seat Number: {self.seat_number}\n"
                f"Price: ${self.price:.2f}\n"
                f"Reservation Status: {reservation_status}\n")

    def __str__(self):
        """
        Provide a string representation of the ticket.

        :return: str, formatted ticket information
        """
        return self.display_ticket_info()

# Example usage:
if __name__ == "__main__":
    ticket1 = Ticket("T123", "Concert", "2024-07-15", "Stadium", "A10", 150.00)
    print(ticket1)

    # Reserving the ticket
    if ticket1.reserve_ticket():
        print("Ticket reserved successfully")
    else:
        print("Ticket is already reserved")
    print(ticket1)

    # Canceling the reservation
    if ticket1.cancel_reservation():
        print("Reservation canceled successfully")
    else:
        print("Ticket is not reserved")
    print(ticket1)
```

```python
class ShoppingCart:
    def __init__(self):
        """
        Initialize a new ShoppingCart instance with an empty list of items.
        """
        self.items = []

    def add_item(self, item):
        """
        Adds an item to the shopping cart by appending it to the list of items.

        :param item: str, the item to add to the cart
        """
        self.items.append(item)

    def remove_item(self, item):
        """
        Removes an item from the shopping cart if it exists in the list.

        :param item: str, the item to remove from the cart
        :return: bool, True if the item was successfully removed, False if the item was not found
        """
        if item in self.items:
            self.items.remove(item)
            return True
        return False

    def view_cart(self):
        """
        Displays the items currently present in the shopping cart.

        :return: str, a formatted string of items in the cart
        """
        if not self.items:
            return "The shopping cart is empty."
        return "Items in the shopping cart:\n" + "\n".join(self.items)

    def clear_cart(self):
        """
        Clears all items from the shopping cart by reassigning an empty list to the items attribute.
        """
        self.items = []

    def __str__(self):
        """
        Provide a string representation of the shopping cart.

        :return: str, a formatted string of items in the cart
        """
        return self.view_cart()

# Example usage:
if __name__ == "__main__":
    cart = ShoppingCart()
    cart.add_item("Laptop")
    cart.add_item("Smartphone")
    print(cart)

    # Removing an item
    if cart.remove_item("Smartphone"):
        print("Smartphone removed from the cart")
    else:
        print("Smartphone not found in the cart")
    print(cart)

    # Viewing the cart
    print(cart.view_cart())

    # Clearing the cart
    cart.clear_cart()
    print("Cart cleared")
    print(cart)
```

```python
10. Imagine a school management system. You have to design the "Student" class using
OOP concepts.The “Student” class has the following attributes:
a. name: Represents the name of the student.
b. age: Represents the age of the student.
c. grade: Represents the grade or class of the student.
d. student_id: Represents the unique identifier for the student.
e. attendance: Represents the attendance record of the student.
The class should also include the following methods:
a. update_attendance(self, date, status): Updates the attendance record of the
student for a given date with the provided status (e.g., present or absent).
b. get_attendance(self): Returns the attendance record of the student.
c. get_average_attendance(self): Calculates and returns the average
attendance percentage of the student based on their attendance record.
```

```python
class Student:
    def __init__(self, name, age, grade, student_id):
        """
        Initialize a new Student instance with the provided attributes.

        :param name: str, the name of the student
        :param age: int, the age of the student
        :param grade: str, the grade or class of the student
        :param student_id: str, the unique identifier for the student
        """
        self.name = name
        self.age = age
        self.grade = grade
        self.student_id = student_id
        self.attendance = {}

    def update_attendance(self, date, status):
        """
        Updates the attendance record of the student for a given date with the provided status.

        :param date: str, the date for which attendance is being updated (e.g., "2024-06-11")
        :param status: str, the status of the student for the given date (e.g., "present", "absent")
        """
        self.attendance[date] = status

    def get_attendance(self):
        """
        Returns the attendance record of the student.

        :return: dict, a dictionary representing the attendance record where keys are dates and values are statuses
        """
        return self.attendance

    def get_average_attendance(self):
        """
        Calculates and returns the average attendance percentage of the student.

        :return: float, the average attendance percentage
        """
        if not self.attendance:
            return 0.0
        
        total_days = len(self.attendance)
        present_days = sum(1 for status in self.attendance.values() if status == "present")
        average_attendance = (present_days / total_days) * 100
        return average_attendance

    def __str__(self):
        """
        Provide a string representation of the student.

        :return: str, a formatted string of student information
        """
        return f"Student Name: {self.name}\nAge: {self.age}\nGrade: {self.grade}\nStudent ID: {self.student_id}"

# Example usage:
if __name__ == "__main__":
    # Creating a student instance
    student1 = Student("John Doe", 15, "10th Grade", "S12345")

    # Updating attendance
    student1.update_attendance("2024-06-10", "present")
    student1.update_attendance("2024-06-11", "absent")
    student1.update_attendance("2024-06-12", "present")

    # Displaying attendance record
    print(f"{student1.name}'s Attendance Record:")
    attendance_record = student1.get_attendance()
    for date, status in attendance_record.items():
        print(f"Date: {date} - Status: {status}")

    # Calculating average attendance
    average_attendance = student1.get_average_attendance()
    print(f"\nAverage Attendance: {average_attendance:.2f}%")

    # Displaying student information
    print("\nStudent Information:")
    print(student1)
```
